```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
#read data
sleep_data = pd.read_csv('/Users/ginabarbagallo/Downloads/workspace/sleep_health_data.csv')

sleep_data.head()
```

```python
#missing values for sleep disorder 
sleep_data.isna()
sleep_data['Sleep Disorder'] = sleep_data['Sleep Disorder'].fillna(0)

sleep_data.head()
```

```python
#variable for model
sleep_data['Disorder Binary'] = np.where(sleep_data['Sleep Disorder'] == 0, 0, 1)
```

```python
#check for class imbalances 
class_counts = sleep_data['Disorder Binary'].value_counts()
print(class_counts)
#data is balanced from 1 and 0s
```

```python
#one hot encoding for categorical variables
sleep_data['Gender'] = np.where(sleep_data['Gender'] == 'Male', 1, 0) #male is 0, female 1

#bmi categories
sleep_data['BMI Category'] = np.where(sleep_data['BMI Category'] == 'Normal', 1, sleep_data['BMI Category']) #normal is 1
sleep_data['BMI Category'] = np.where(sleep_data['BMI Category'] == 'Overweight', 2, sleep_data['BMI Category']) #overweight is 2
sleep_data['BMI Category'] = np.where(sleep_data['BMI Category'] == 'Obese', 3, sleep_data['BMI Category']) #obese is 3

#jobs
sleep_data['Occupation'] = np.where(sleep_data['Occupation'] == 'Nurse', 1, sleep_data['Occupation']) #nurse is 1
sleep_data['Occupation'] = np.where(sleep_data['Occupation'] == 'Doctor', 2, sleep_data['Occupation']) #doctor is 2
sleep_data['Occupation'] = np.where(sleep_data['Occupation'] == 'Software Engineer', 3, sleep_data['Occupation']) #soft eng is 3
sleep_data['Occupation'] = np.where(sleep_data['Occupation'] == 'Sales Representative', 4, sleep_data['Occupation']) #sales rep is 4
sleep_data['Occupation'] = np.where(sleep_data['Occupation'] == 'Teacher', 5, sleep_data['Occupation']) #teacher is 5
sleep_data['Occupation'] = np.where(sleep_data['Occupation'] == 'Accountant', 6, sleep_data['Occupation']) #accountant is 6
sleep_data['Occupation'] = np.where(sleep_data['Occupation'] == 'Lawyer', 7, sleep_data['Occupation']) #lawyer is 7
sleep_data['Occupation'] = np.where(sleep_data['Occupation'] == 'Scientist', 8, sleep_data['Occupation']) #scientist is 8
sleep_data['Occupation'] = np.where(sleep_data['Occupation'] == 'Manager', 9, sleep_data['Occupation']) #manager is 9
sleep_data['Occupation'] = np.where(sleep_data['Occupation'] == 'Engineer', 10, sleep_data['Occupation']) #engineer is 10

sleep_data['Occupation'] = np.where(sleep_data['Occupation'] == 'Salesperson', 4, sleep_data['Occupation']) 
```

```python
occupations = sleep_data['Occupation'].value_counts()
print(occupations)

#drop 3,8,9 not enough observations for each category
sleep_data = sleep_data[~sleep_data["Occupation"].isin([3, 8, 9])]

#check the drop was done correctly
occupations = sleep_data['Occupation'].value_counts()
print(occupations)
```

```python
#check blood pressure categories
blood_pressures = sleep_data['Blood Pressure'].value_counts()
print(blood_pressures)
```

```python
#splitting the blood pressures into two so that both are numeric and can be used in the RF model
# Split into systolic and diastolic
sleep_data[['systolic', 'diastolic']] = sleep_data['Blood Pressure'].str.split('/', expand=True)

# Convert to numeric
sleep_data['systolic'] = pd.to_numeric(sleep_data['systolic'])
sleep_data['diastolic'] = pd.to_numeric(sleep_data['diastolic'])
```

```python
sleep_data.info()
```

```python
sleep_data.head()
```

```python
#Drop blood pressure, drop sleep disorder 
sleep_data = sleep_data.drop(columns = ['Sleep Disorder'])
sleep_data = sleep_data.drop(columns = ['Blood Pressure'])
sleep_data.head()
```

```python
#turn occupation and bmi category to numeric 
sleep_data['Occupation'] = pd.to_numeric(sleep_data['Occupation'])
sleep_data['BMI Category'] = pd.to_numeric(sleep_data['BMI Category'])
```

```python
#train - test split
y = sleep_data['Disorder Binary'] 

X = sleep_data.drop(columns = ['Disorder Binary', 'Person ID'])

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3)

#RANDOM FOREST MODEL
rf_sleep = RandomForestClassifier(n_estimators=100, random_state=42, max_depth=10)
rf_sleep.fit(X_train, y_train)
y_pred = rf_sleep.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# Confusion Matrix
print(confusion_matrix(y_test, y_pred))

# Detailed Classification Report
print(classification_report(y_test, y_pred))
```

```python
# Feature Importance
feature_importances = rf_sleep.feature_importances_

# Create a DataFrame with feature names and their importance
feature_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': feature_importances
})

# Sort the features by importance
feature_df = feature_df.sort_values(by='Importance', ascending=False)

# Display the feature importance
print(feature_df)
```
